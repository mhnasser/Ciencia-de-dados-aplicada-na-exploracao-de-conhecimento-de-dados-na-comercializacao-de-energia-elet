```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Sudeste

```python
df = pd.read_excel('CMO_relatorio.xlsx', sheet_name=0)
df
```

```python
# Melt the dataframe
df_melted = df.reset_index().melt(id_vars='Data', value_vars=df.columns, var_name='din_instante', value_name='cmo')
df_melted.head()
```

```python
df_melted_SE = df_melted
df_melted_SE['din_instante'] = df_melted_SE['Data'] +' '+df_melted_SE['din_instante']+':00'
df_melted_SE['id_subsistema'] = 'SE'
df_melted_SE.drop('Data', axis=1, inplace=True)
df_melted_SE
```

## Nordeste

```python
# Melt the dataframe
df_melted = df.reset_index().melt(id_vars='Data', value_vars=df.columns, var_name='din_instante', value_name='cmo')
df_melted.head()
```

## Sul

```python
df = pd.read_excel('CMO_relatorio.xlsx', sheet_name=3)
df
```

```python
# Melt the dataframe
df_melted = df.reset_index().melt(id_vars='Data', value_vars=df.columns, var_name='din_instante', value_name='cmo')
df_melted.head()
```

```python
df_melted_S = df_melted
df_melted_S['din_instante'] = df_melted_S['Data'] +' '+df_melted_S['din_instante']+':00'
df_melted_S['id_subsistema'] = 'S'
df_melted_S.drop('Data', axis=1, inplace=True)
df_melted_S
```

```python
df_final = pd.concat([df_melted_NE,
                      df_melted_NO,
                      df_melted_S,
                      df_melted_SE], axis=0, ignore_index=True)

df_final

df_final.to_csv('cmo-semi-horario.csv',index=False)
```

```python
# Assuming you've already created df_final and added 'din_instante'
df_final['din_instante'] = pd.to_datetime(df_final['din_instante'], dayfirst=True)
df_final['din_instante'] = df_final['din_instante'].dt.floor('H')

# Drop duplicate rows based on ['id_subsistema', 'din_instante']
df_final = df_final.drop_duplicates(subset=['id_subsistema', 'din_instante']).reset_index(drop=True)

df_final
```

```python
df_final['cmo'] = df_final['cmo'].apply(lambda x: str(x).replace(',','.'))
df_final['cmo'] = df_final['cmo'].astype(float)
```

```python
# Setting the style for seaborn
sns.set_style("whitegrid")

# Read the data (assuming your data is in a CSV file)
data = df_final

# Convert 'din_instante' column to datetime if it's not
data['din_instante'] = pd.to_datetime(data['din_instante'])

# Plotting with hue for 'id_subsistema'
plt.figure(figsize=(12, 7))
sns.lineplot(x='din_instante', y='cmo', hue='id_subsistema', data=data, markersize=4, palette="tab10")

plt.title('Temporal Plot of cmo vs. din_instante')
plt.xlabel('din_instante')
plt.ylabel('cmo')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
